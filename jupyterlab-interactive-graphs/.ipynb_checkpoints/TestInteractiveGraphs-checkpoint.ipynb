# Test `jupyterlab-interactive-graphs`

Run the cells below. You should get **hover tooltips**, **scroll zoom**, and **box/lasso select** (toolbar) on Plotly figures.

If you see a static image only, confirm the extension is enabled: **Help → About JupyterLab** or run `jupyter labextension list` for `jupyterlab-interactive-graphs`.

```python
import plotly.express as px
import plotly.graph_objects as go

df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", title="Iris — hover & select")
fig.show()
```

```python
# Line + markers (select points on the line)
fig2 = go.Figure(
    go.Scatter(
        x=list(range(20)),
        y=[(i % 5) ** 2 for i in range(20)],
        mode="lines+markers",
        name="demo",
    )
)
fig2.update_layout(title="Line + markers")
fig2.show()
```

```python
# Bar chart (hover on bars)
fig3 = px.bar(x=["A", "B", "C"], y=[3, 1, 4], title="Bars — hover")
fig3.show()
```
