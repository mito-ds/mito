# Test `jupyterlab-interactive-graphs`

The first cell calls **`configure()`**: Plotly uses the JupyterLab MIME renderer; Matplotlib uses the **ipympl** `widget` backend (install **`pip install 'jupyterlab-interactive-graphs[matplotlib]'`** first).

- **Plotly**: hover, scroll zoom, box/lasso select from the mode bar.
- **Matplotlib**: interactive canvas with pan/zoom toolbar (not the same as Plotly’s lasso, but fully usable in the notebook).

Confirm the Plotly lab extension: `jupyter labextension list` → `jupyterlab-interactive-graphs` enabled.

```python
from jupyterlab_interactive_graphs import configure

# Plotly → jupyterlab renderer; Matplotlib → ipympl widget backend (needs: pip install 'jupyterlab-interactive-graphs[matplotlib]')
configure()

import plotly.express as px
import plotly.graph_objects as go

df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", title="Iris — hover & select")
fig.show()
```

```python
# Line + markers (select points on the line)
fig2 = go.Figure(
    go.Scatter(
        x=list(range(20)),
        y=[(i % 5) ** 2 for i in range(20)],
        mode="lines+markers",
        name="demo",
    )
)
fig2.update_layout(title="Line + markers")
fig2.show()
```

```python
# Bar chart (hover on bars)
fig3 = px.bar(x=["A", "B", "C"], y=[3, 1, 4], title="Bars — hover")
fig3.show()
```

```python
# Matplotlib — interactive canvas (pan/zoom toolbar) via ipympl
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(np.linspace(0, 6, 80), np.sin(np.linspace(0, 6, 80)), "o-", markersize=3)
ax.set_title("Matplotlib — use toolbar to pan/zoom")
plt.show()
```
