Run **all cells** (first cell runs `configure()`). You should see a **Mito / Graph** strip beside each chart output.

## 1 — Plotly scatter

```python
import plotly.express as px
df = px.data.iris()
px.scatter(df, x="sepal_width", y="sepal_length", color="species", title="Plotly: Iris").show()
```

## 2 — Matplotlib (ipympl widget)

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(5, 3))
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label="sin")
ax.plot(x, np.cos(x), label="cos")
ax.legend()
ax.set_title("Matplotlib: sin & cos")
plt.show()
```

## 3 — Plotly bar chart

```python
import plotly.express as px
px.bar(x=["A", "B", "C", "D"], y=[2, 5, 3, 7], title="Plotly: bar").show()
```

## 4 — PNG image output (matplotlib → buffer)

Tests the **image/png** path for the Mito strip.

```python
import io
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image, display

fig, ax = plt.subplots(figsize=(4, 3))
ax.imshow(np.random.RandomState(0).rand(12, 12), cmap="viridis")
ax.set_title("Matplotlib heatmap → PNG")
buf = io.BytesIO()
fig.savefig(buf, format="png", bbox_inches="tight")
plt.close(fig)
display(Image(data=buf.getvalue()))
```

## 5 — Plotly line chart

```python
import plotly.graph_objects as go
go.Figure(
    data=[go.Scatter(y=[1, 3, 2, 4, 3, 5], mode="lines+markers", name="series")],
    layout=go.Layout(title="Plotly: line graph"),
).show()
```
